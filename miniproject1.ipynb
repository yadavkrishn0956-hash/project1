# Converting Date Column
(Converting the Date column into a datetime object and sorting the data by date )

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

# Plotting Closing Prices
( visualizing Samsung's closing prices over time to observe overall trends and fluctuations. The dates are rotated for better readability )

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], color='red')
plt.title('Samsung Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

# Plotting Open vs Close Prices

( comparing Samsung's opening and closing prices over time to see how the stock moves within each trading day. This helps identify daily trends and the relationship between opening and closing values )

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label='Close', color='green', linestyle='-', marker='o', markersize=3)
plt.plot(df['Date'], df['Open'], label='Open', color='red', linestyle='-', marker='x', markersize=1)

plt.title('Open and Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

# Calculating and Plotting Daily Price Change

( calculating the daily price change by subtracting the opening price from the closing price. This plot shows how much the stock price moved each day, helping us understand daily volatility )

```python
df['Daily_Change'] = df['Close'] - df['Open']


import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Daily_Change'], color='purple')
plt.xlabel('Date')
plt.ylabel('Daily Change')
plt.title('Daily Price Change')
plt.show()
```

# Calculating and Plotting Daily Percentage Change

( calculating the daily percentage change to see how much the stock price moves relative to its opening price each day. This plot helps us understand the magnitude of daily fluctuations in percentage terms.\)

```python
df['Pct_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100

# Plot Percentage Change
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Pct_Change'], color='green')
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.title('Daily Percentage Change')
plt.show()
```

# Calculating and Plotting Daily Percentage Change

( calculating the daily percentage change by comparing the closing price with the opening price. This plot visualizes the daily price movements in percentage terms, helping us spot high-volatility days at a glance )

```python
df['Pct_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100

# Plot Percentage Change
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Pct_Change'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.title('Daily Percentage Change')
plt.show()
```
